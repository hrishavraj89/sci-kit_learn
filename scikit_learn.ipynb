```python
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
load_iris()
```

```python
load_iris(return_X_y = True)
```

```python
X, y = load_iris(return_X_y = True)
```

```python
#all input variables
X
```

```python
#all output variables
y
```

all input and output features have been separated sucessfullyin X and y  variables.

```python
# imported the Linear Regression package
from sklearn.linear_model import LinearRegression
```

```python
# creating the model. This is the phase where model is not learning anything, we have just created it.
Model = LinearRegression()
```

```python
Model.fit(X, y)
```

```python
Model.predict(X)
```

```python
pred1 = Model.predict(X)
```

```python
plt.scatter(pred1, y)
```

```python
from sklearn.neighbors import KNeighborsRegressor
mod = KNeighborsRegressor()
```

```python
mod.fit(X, y)
```

```python
mod.predict(X)
```

```python
pred2 = mod.predict(X)
```

```python
plt.scatter(pred2, y)
```
