```python
from sklearn.datasets import fetch_openml
```

```python
df = fetch_openml('titanic', version = 1, as_frame = True)['data']
```

```python
df.info()
```

```python
#finding missing value
df.isnull()
```

```python
# a better approach for finding null values
miss_val = df.isnull().sum()
miss_val
```

```python
sns.set()
miss_val_per = (pd.DataFrame(miss_val / len(df))*100)
miss_val_per.plot(kind = 'bar', title = 'Missing values in percentage', ylabel = 'Percentage')
```

```python
print(f"The size of the data is {df.shape}")
```

```python
df.drop(['body'], axis = 1, inplace = True)
# inplace = True is used the modify the changes to the main dataset
print(f"The size of the data after dropping the feature is {df.shape}")
```
